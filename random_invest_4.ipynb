### Setup

```python
path = untar_data(URLs.MNIST_TINY)
df = build_df(path)

y_names = ['point_topleft_x', 'point_topleft_y',]

db =   DataBlock(blocks=(ImageBlock(cls=PILImageBW), 
                         PointBlock), 
                splitter=RandomSplitter(seed=0),
                get_x=ColReader('fn', pref=path),
                get_y=ColReader(y_names),
                )
```

```python
net3 = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=784, out_features=28),
            nn.Linear(in_features=28, out_features=2, bias=False),
            SigmoidRange(-1., 1),
            )
```

```python
relu3 = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=784, out_features=28),
            nn.ReLU(),
            nn.Linear(in_features=28, out_features=2, bias=False),
            SigmoidRange(-1., 1),
            )
```

```python
relu4 = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=784, out_features=28),
            nn.ReLU(),
            nn.Linear(in_features=28, out_features=28),
            nn.ReLU(),
            nn.Linear(in_features=28, out_features=2, bias=False),
            SigmoidRange(-1., 1),
            )
```

```python
net5 = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=784, out_features=56),
            nn.Linear(in_features=56, out_features=28),
            nn.Linear(in_features=28, out_features=2, bias=False),
            SigmoidRange(-1., 1),
            )
```

```python
dl = db.dataloaders(df)

learn_net3 = Learner(
                dl, 
                net3,
                loss_func=MSELossFlat(), 
                metrics=[mae, R2Score()],
                #cbs = CSVLogger('tmp.csv'),
                )

learn_net3.fit(20)
```

```python
dl = db.dataloaders(df)

learn_relu3 = Learner(
                dl, 
                relu3,
                loss_func=MSELossFlat(), 
                metrics=[mae, R2Score()],
                #cbs = CSVLogger('tmp.csv'),
                )

learn_relu3.fit(20)
```

```python
dl = db.dataloaders(df)

learn_relu4 = Learner(
                dl, 
                relu4,
                loss_func=MSELossFlat(), 
                metrics=[mae, R2Score()],
                #cbs = CSVLogger('tmp.csv'),
                )

learn_relu4.fit(20)
```

```python
dl = db.dataloaders(df)

learn_net5 = Learner(
                dl, 
                net5,
                loss_func=MSELossFlat(), 
                metrics=[mae, R2Score()],
                #cbs = CSVLogger('tmp.csv'),
                )

learn_net5.fit(20)
```
